**PCA**

The Principal Component Analysis is a popular unsupervised learning technique for reducing the dimensionality of data. It increases interpretability yet, at the same time, it minimizes information loss. It helps to find the most significant features in a dataset and makes the data easy for plotting in 2D and 3D.

Applications of PCA in Machine Learning

*   PCA is used to visualize multidimensional data.
*   It is used to reduce the number of dimensions in healthcare data.

*   PCA can help resize an image.
*   It can be used in finance to analyze stock data and forecast returns.

*   PCA helps to find patterns in the high-dimensional datasets.






```python
import pandas as pd 
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
uni = pd.read_csv("Copy of Universities.csv")
uni.head()
```

```python
# Considering only numerical data 
uni1 = uni.iloc[:,1:]
uni1.head()
```

```python
# Normalizing the numerical data 
std = StandardScaler()
uni_normal=std.fit_transform(uni1)
uni_normal
```

```python
pca2=PCA(n_components=2)
pca_components=pd.DataFrame(pca2.fit_transform(uni_normal),columns=['a','b'])
pca_components.head(7)
```

```python
plt.scatter(pca_components.a,pca_components.b)
plt.show()
```

```python
pca3 = PCA(n_components=3)
pca3.fit(uni_normal)
```

```python
pca = PCA(n_components = 6)
pca_values = pca.fit_transform(uni_normal)
```

```python
pca_values
```

```python
# The amount of variance that each PCA explains is 
var = pca.explained_variance_ratio_
var
```

```python
# Cumulative variance 
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
pca.components_
```

```python
# Variance plot for PCA components obtained  Scree plot
plt.plot(var1,color="red")
plt.title('Scree plot')
plt.show()
```
